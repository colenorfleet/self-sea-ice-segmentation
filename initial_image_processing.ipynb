```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './dataset_cole/reconstructed_rgb/0019.jpg'

image = cv2.imread(path)
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
closing = cv2.morphologyEx(img_gray, cv2.MORPH_CLOSE, kernel)
```

```python
ret, thresh = cv2.threshold(closing, 5, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

image_copy = image.copy()
#thickness = cv2.FILLED
cv2.drawContours(image=image_copy, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)

plt.imshow(image_copy)
```

```python
second_gray = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

second_closing = cv2.morphologyEx(second_gray, cv2.MORPH_CLOSE, kernel)

f, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(second_gray, cmap='gray')
ax[1].imshow(second_closing, cmap='gray')

plt.show()
```

```python
ret2, thresh2 = cv2.threshold(second_closing, 5, 255, cv2.THRESH_BINARY)
plt.imshow(thresh2, cmap='gray')
```

```python
contours_2, hierarchy_2 = cv2.findContours(thresh2, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

second_image_copy = image.copy()
third_image_copy = image.copy()
#thickness = cv2.FILLED
cv2.drawContours(image=second_image_copy, contours=contours_2, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
cv2.drawContours(image=third_image_copy, contours=contours_2, contourIdx=-1, color=(0, 255, 0), thickness=cv2.FILLED, lineType=cv2.LINE_AA)


f, axes = plt.subplots(1, 2, figsize=(20, 10))
axes[0].imshow(second_image_copy)
axes[1].imshow(third_image_copy)
```
